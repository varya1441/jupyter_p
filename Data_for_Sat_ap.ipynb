```python
import requests
from bs4 import BeautifulSoup
import csv 
import pandas as pd
import numpy as np
```

```python
a='container panel-row-style panel-row-style-for-2310-4'
q=[]
for i in [3,4,5,6]:
    a=a[:-1] + str(i)
    q.append(a) 
```

```python
df=pd.DataFrame()
```

```python
URL = "https://www.isispace.nl/products/"
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content, 'html5lib') 
  
quotes=[]  # a list to store quotes 
for i, el in enumerate(q):
    table = soup.find('div', attrs = {'class':[q[i]]}) 
    for row in table.findAll('div'): 
        quote = {} 
        quote['product-text'] = row.text.replace('\n', '')
        quote['url'] = row.a['href'] 
        quote['img'] = row.img['src'] 
        quotes.append(quote) 
    df=df.append(quotes)    
#filename = 'inspirational_quotes.csv'
#with open(filename, 'wb') as f: 
 #   w = csv.DictWriter(f,['product-text','url','img']) 
    #w.writeheader() 
   
       # w.writerow(quote) 
```

```python
df = df.drop_duplicates(subset='product-text', keep='first')
df = df[df['product-text'].map(len) > 0]
```

```python
ind=[x for x in range(0,len(df))]
df.index=ind
```
